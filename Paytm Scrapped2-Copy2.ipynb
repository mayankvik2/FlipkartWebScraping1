```python
import requests
import re
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
from collections import OrderedDict
```

```python
#laptop
base_url1="https://paytm.com/shop/search?q=laptop&from=organic&child_site_id=1&site_id=1&category=6453&availability=1&brand=2562%2C1758%2C1707"
#tablet
#base_url2="https://paytm.com/shop/search?q=samsung%20tablet&from=organic&child_site_id=1&site_id=1&category=66782&brand=2578"
base_url2="https://paytm.com/shop/search?q=tablet&from=organic&child_site_id=1&site_id=1&category=66782&brand=2578%2C439603&availability=1"
#smartphone
base_url3="https://paytm.com/shop/search?q=samsung%20smartphone&from=organic&child_site_id=1&site_id=1&category=66781&brand=2578%2C2565&availability=1"

input1=input("Enter the electronic product you want to purchase:")
if input1=="smartphone":
    base_url=base_url3
elif input1=="tablet":
    base_url=base_url2
elif input1=="laptop":
    base_url=base_url1
else:
    print("Only Laptop Tablet Smartphone products are avilable")

print(base_url)    
```

```python
r=requests.get(base_url)
soup = BeautifulSoup(r.text,"html5lib")
print(soup.title)

all_links = [link.get('href') for link in soup.findAll(href=True) if 'https://paytm.com/shop' in  str(link.get('href'))]
all_links.pop(0)  
print(all_links[0]) 
```

```python
final_df = pd.DataFrame()
for url in all_links:
        r=requests.get(url)
        soup1 = BeautifulSoup(r.text,"html5lib")
        try:
            #Offer
            Offer=[]
            for div in soup1.find_all(class_='_3M3B'):
                temp=div.text
                Offer.append(temp)
            #Price
            price=[]
            for div in soup1.find_all(class_='_1y_y'):
                temp=div.text
                temp2=re.findall('\d+',temp)
                price.append(temp2)
            c_price=price[0][0]+price[0][1]
            #Descrption
            Des=[]
            for div in soup1.find_all(class_='_2LOI'):
                temp=div.text
                Des.append(temp)
            #Categories                
            Col=[]
            for div in soup1.find_all(class_='w3LC'):
                temp=div.text
                Col.append(temp)
            #dic=OrderedDict({'Description':Des})
            #dic_name=OrderedDict({'Category':Col})
            dictionary = dict(zip(Col,Des))
            dic_final=dict({'Price':c_price,'Offer':Offer,"URL":url})
            dic_final.update(dictionary)
            #dic_final.update(dic)
            final_df = final_df.append(dic_final,ignore_index=True)
            print(final_df) 

        except Exception as e:
            print("Incorrect url")

            
```

```python

final_df.to_excel("InpectPaytmLaptop.xlsx")
```
